```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv('MentalHealthSurvey.csv')

# Summary statistics for CGPA, Sleep Patterns, and Stress-Related Attributes
cgpa_summary = df['cgpa'].describe()
sleep_summary = df['average_sleep'].describe()
stress_summary = df[['academic_pressure', 'financial_concerns', 'future_insecurity']].describe()

cgpa_summary, sleep_summary, stress_summary
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["average_sleep"], df["cgpa"], color='blue', alpha=0.08)
plt.title("Sleep vs. Academic Performance (CGPA)")
plt.xlabel("Average Sleep per Night (hours)")
plt.ylabel("CGPA")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["sports_engagement"], df["academic_pressure"], color='green', alpha=0.08)
plt.title("Sports Engagement vs. Stress Relief")
plt.xlabel("Sports per Week")
plt.ylabel("Academic Pressure Score")
plt.grid(True)
plt.show()
```

```python
time_period = range(len(df))  # Replace with your actual time period data (e.g., semesters, months)
plt.figure(figsize=(10, 5))
plt.plot(time_period, df["average_sleep"], label='Sleep Hours', marker='o')
plt.plot(time_period, df["cgpa"], label='CGPA', marker='s')
plt.title("Sleep and CGPA Over Time")
plt.xlabel("Students")
plt.ylabel("Values")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Histograms 
plt.figure(figsize=(8, 5))
plt.hist(df["average_sleep"], bins=5, color='purple', alpha=0.7)
plt.title("Histogram of Sleep Hours")
plt.xlabel("Sleep Hours")
plt.ylabel("Frequency")
plt.show()
```
